```python
import requests
from bs4 import BeautifulSoup
import json

url = "https://www.nytimes.com/books/best-sellers/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

books = soup.find_all("div", class_="css-v2kl5d")


best_seller = []

for i in range(len(books)-2):
    
    category = books[i+1].find_all("a", rel="canonical")[0].text.strip()
    
    books_list = []
    
    for j in range(5):
        
        title = books[i+1].find_all("h3")[j].text.strip() if books[i+1].find_all("h3")[j].text.strip() else "N/A"
        position = int(j+1)
        description = books[i+1].find_all("p", itemprop="description")[j].text.strip() if books[i+1].find_all("p", itemprop="description")[j].text.strip() else "N/A"
        image_url = books[i+1].find_all("img")[j].get("src") if books[i+1].find_all("img")[j].get("src") else "N/A"
    
        book =  {"title": title, 
                 "position": position, 
                 "description": description, 
                 "img_url": image_url}
        
        books_list.append(book)
    
    

    dic = {"week": 1, 
       "data": [
           {"category": category, 
            "books": books_list
           }
       ]
      }
    
    
    best_seller.append(dic)


for k in range(103):
    
    url = "https://www.nytimes.com" + str(books[0].find_all("div", class_="css-iouqpc")[0].find("a").get("href"))
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    books = soup.find_all("div", class_="css-v2kl5d")
    
    
    for i in range(len(books)-2):
    
        category = books[i+1].find_all("a", rel="canonical")[0].text.strip()

        books_list = []

        for j in range(5):

            title = books[i+1].find_all("h3")[j].text.strip() if books[i+1].find_all("h3")[j].text.strip() else "N/A"
            position = int(j+1)
            description = books[i+1].find_all("p", itemprop="description")[j].text.strip() if books[i+1].find_all("p", itemprop="description")[j].text.strip() else "N/A"
            image_url = books[i+1].find_all("img")[j].get("src") if books[i+1].find_all("img")[j].get("src") else "N/A"

            book =  {"title": title, 
                     "position": position, 
                     "description": description, 
                     "img_url": image_url}

            books_list.append(book)



        dic = {"week": k+2, 
           "data": [
               {"category": category, 
                "books": books_list
               }
           ]
          }


        best_seller.append(dic)
     
    
    

with open("data.json", "w") as json_file:
    json.dump(best_seller, json_file)
```
